```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "a": [1.0, np.nan, np.nan, 4.0],
    "b": [2.0, np.nan, 3.0,   np.nan],
    "c": [3.0, np.nan, 4.0,   5.0],
})
print(df)
```

```python

# 1. how="any" (디폴트) → 하나라도 결측치면 제거
df.dropna(how="any")
```

```python

# 2. how="all" → 모두 결측치일 때만 제거
df.dropna(how="all")
```

```python
# 3. thresh=2 → 결측치가 아닌 값이 2개 이상인 행만 유지
# threshold -> 임계치
df.dropna(thresh=2)
```

```python
# 4. subset → 특정 컬럼만 기준으로 삼기
# a 컬럼 기준, a가 NaN인 행만 제거
# subaet ->  부분집합
df.dropna(subset=["a"])
```

```python
df
```

```python
# 5. axis=1 → 행이 아닌 열 기준으로 제거
df.dropna(axis=0, how="any")
#2차원은 shape이 2개임 
```

```python
import pandas as pd

data = {
    'Math':[90,80],
    'English':[85,95],
    'Science':[70,75]
}

df = pd.DataFrame(data, index= ['Student_a','Student_b'])
print(df)
```

```python
#과목별 총점 
col_sum = df.sum(axis=0)
print(col_sum)
```

```python
#학생별 전과목 총점
row_sum = df.sum(axis=1)
print(row_sum)
```

```python
import pandas as pd
import numpy as np 

data = {
    'A':[1,2,np.nan],
    'B':[4,np.nan,np.nan],
    'C':[7,8,9]
}

df = pd.DataFrame(data, index=['Row1','Row2','Row3'])
print(df)
```

```python
#각 열(column)별 데이터 개수
count_col = df.count(axis=0)
print(count_col)
```

```python
#각 행(Row)별 데이터 개수
count_row = df.count(axis=1)
print(count_row)
```

```python
import pandas as pd
import numpy as np 

data = {
    'Math':[90,np.nan,80],
    'English':[80,70,np.nan],
    'Science':[100,100,100]
}

df = pd.DataFrame(data, index= ['Student_a','Student_b','Student_c'])
print(df)
```

```python
#과목별 평균 (수학 2명의 평균 , 영어도 2명의 평균)
mean_by_subject = df.mean(axis=0)
print(mean_by_subject)
```

```python
#학생별평균 본인이 응시한 과복의 평균
mean_by_student =df.mean(axis=1)
print(mean_by_student)
```

```python
#min, max 과목별
min_subject = df.min(axis=0)
print(min_subject)
print("--------------------------------------------")
max_subject = df.max(axis=0)
print(max_subject)
```

```python
##min, max 학생별
min_student = df.min(axis=1)
print(min_student)
print("--------------------------------------------")
max_student = df.max(axis=1)
print(max_student)
```
