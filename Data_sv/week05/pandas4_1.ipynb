```python
# list comprehension
# 파이썬 복습

# 1. 기존 리스트의 모든 요소에 특정 연산을 적용하여 새 리스트를 만드는 경우
squares = [x**2 for x in range(10)] # [■ for ■ in ◆] 0부터 9까지를 제곱해서 새 리스트에 집어넣어라
print(squares)

words = ['python', 'is', 'awesome']
upper_words = [w.upper() for w in words]
print(upper_words)
#---------------------------------------
# 2. 조건에 맞는 요소만 골라내고 나머지는 버림
evens = [x for x in range(1, 21) if x % 2 == 0] # [■ for ■ in ◆ if ●]
print(evens)

fruits = ['apple', 'banana', 'cherry', 'kiwi', 'mango']
a_fruits = [f for f in fruits if 'a' in f]
print(a_fruits)
#---------------------------------
# 3. 조건에 의해서 새로운 리스트 만들기
numbers = [1, 2, 3, 4, 5]
labels = ['Even' if x % 2 == 0 else 'Odd' for x in numbers] # [♣ if ■ for ■ in ◆] # 이거 이해 못함 저 클로버 위치
# for x in numbers 이 x를 조건으로(if x % 2 == 0) 사용
print(labels)
#---------------------
# 4. for안에 또 다른 for
matrix = [[1, 2], [3, 4], [5, 6]]
flat = [num for row in matrix for num in row]

print(flat)
```

```python
matrix = [[1, 2], [3, 4], [5, 6]]

for row in matrix:
    for num in row:
        print(num)

# 두 for를 한 줄에 쓰면
# for row in matrix for num in row

# 마지막 num을 앞쪽에 < 왜그런데요
# num for row in matrix for num in row
```

```python
# 5. dictionary comprehension

square_dict = {x: x**2 for x in range(5)}
print(square_dict)
```

```python
# 행만 있냐? 컬럼도 있다
# 멀티인덱스를 컬럼도 할 수 있다
row_index = pd.MultiIndex.from_product([[2013, 2014], ['1st', '2nd']], names = ['Year', 'Visit'])
col_index = pd.MultiIndex.from_product([['Bob', 'Guido', 'Sue'], ['HR', 'Temp']], names= ['Patient', 'Data type'])
print(row_index)
print(col_index)
```

```python
data_hr = np.random.randint(50, 80, 12) # 50부터 79까지의 정수중에서 무작위로 12개 선택
data_temp = 37 + np.random.randn(12) # 평균이 0이고 표준편차가 1인 표준정규분포 random number 12개

data_x = [i for i in zip(data_hr, data_temp)]
data_x  
```

```python
arr = np.array(data_x)
print(arr)
```

```python
# Reshape 함수 to 4 rows and 6 cols
# 전체 자료값 자리 개수는 같아야 함.
# shape를 바꿈 원래 (12,2) 였는데... 뭘 하든간에 전부 24개기만 하면 됨.
arr_multi = arr.reshape(4,6)
arr_multi
```

```python
health_data = pd.DataFrame(arr_multi, columns=col_index, index=row_index)
health_data

```
